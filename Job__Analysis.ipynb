```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
```

```python
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome(options=chrome_options)
wait = WebDriverWait(driver, 10)

url = "https://careers.thomsonreuters.com/us/en/search-results?keywords=india"
driver.get(url)

job_links = []
```

```python
# Extract job links from all pages
while True:
    time.sleep(5)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    job_elements = soup.find_all('a', class_='au-target')

    for element in job_elements:
        job_link = element.get('href')
        if job_link:  # Ensure href is not None
            job_links.append(job_link)

    # Navigate to the next page
    try:
        next_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "a.next-btn")))
        next_button.click()
    except:
        break
```

```python
# Links to remove
links_to_remove = {
    "https://careers.thomsonreuters.com/us/en/cookiesettings",
    "#acc-skip-content",
    "/",
    "https://careers.thomsonreuters.com/us/en/jointalentcommunity",
    "https://careers.thomsonreuters.com/us/en/military-and-veterans",
    "https://careers.thomsonreuters.com/us/en/search-results",
    "https://careers.thomsonreuters.com/us/en/jobcart",
    "javascript:void(0);",
    "javascript:void(0)"
}

# Updated criteria: Remove links starting with specific pattern
pattern_to_remove = "https://careers.thomsonreuters.com/us/en/search-results?"

old_job_links = job_links

# Cleaning up the job_links list by removing the specified links
cleaned_job_links = [link for link in job_links if (link not in links_to_remove and not link.startswith(pattern_to_remove))]

job_links = cleaned_job_links
print(len(job_links))
print('-'*120)
```

```python
job_details = {}  # Dictionary to store job link and description

# Visit each job page and extract the description
for link in job_links:
    # Check if the link is a valid URL
    if not link.startswith("http"):
        print("Invalid URL skipped:", link)
        continue

    try:
        driver.get(link)
        time.sleep(5)  # Increase wait time for the page to load

        # Check if the job description section is present
        WebDriverWait(driver, 5).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "section.job-description"))
        )
        detail_soup = BeautifulSoup(driver.page_source, 'html.parser')
        job_description_section = detail_soup.find('section', class_='job-description')
        if job_description_section:
            job_description = job_description_section.get_text(strip=True)
            job_details[link] = job_description
        else:
            print("Job description not found for:", link)
            job_details[link] = "Description not found"
    except Exception as e:
        print("Error loading page:", link, "\nException:", e)
        job_details[link] = "Failed to load page"

# Close the browser
driver.quit()

# Now, job_details dictionary contains all the job descriptions with their respective links
for link, description in job_details.items():
    print(f"Link: {link}\nDescription: {description}\n")
```

```python
# Initialize the OpenAI client with your API key
client = OpenAI(
    api_key="YOUR API KEY HERE",  # Replace with your actual API key
)

def chat_gpt(prompt):
    # Generate a response using the OpenAI API
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}]
    )

    # Extracting the response content correctly
    return response.choices[0].message.content.strip()
```
